```python
l=[-4,5,6,10,23,-10,-3,45]
x=np.array(l)
"""
y=f(x)
{x<0, x**2, x>0 2x+1}
"""
y=np.where(x<0, x**2,2*x+1)
print x
print y
plt.bar(x,y)
plt.show()
```

```python
l=[-4,5,6,0,10,23,-10,-3,0,45]
x=np.array(l)
"""
y=f(x)
{x<0, x**2, x>0 2x+1, x==0 3}
"""
y1=np.where(x<0, x**2,x)
y=np.where(y1>0, 2*y1+1, 3)

print x
print y
plt.bar(x,y)
plt.show()
```
